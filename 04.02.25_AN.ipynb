```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = cv2.imread("text.png", cv2.IMREAD_GRAYSCALE)
data.shape
```

```python
plt.imshow(data, cmap='gray')
```

```python
y, x = data.shape
```

```python
counting = False

letters = []

for c in range(x):

    if np.sum(data[:, c]) > 0 and not counting:
        counting = True
        start = c
    
    if (np.sum(data[:, c]) == 0 or c == x - 1) and counting:
        counting = False
        end = c
        letter = np.zeros((y, end - start), dtype=np.uint8)
        letter = data[:, start: end]
        letters.append(letter)
```

```python
plt.figure(figsize=(10, 10))

for i, letter in enumerate(letters):
    plt.subplot(1, len(letters), i + 1)
    plt.imshow(letter, cmap='gray')
    plt.axis('off')

plt.show()
```
