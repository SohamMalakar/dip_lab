```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file1 = input("Enter a rgb image filename:")
file2 = input("Enter output image filename:")
```

```python
data = plt.imread(file1)
```

```python
rows = int(input("Enter the number of rows:"))
cols = int(input("Enter the number of cols:"))
```

```python
fig = plt.figure(figsize=(20,20))

i = 1

for _ in range(rows):
    for _ in range(cols):
        fig.add_subplot(rows, cols, i)
        plt.axis("off")
        plt.imshow(data)
        i += 1

plt.savefig(file2)
plt.show()
```

```python
y, x = data.shape[:2]
m = y/x
m
```

```python
file3 = input("Enter output image filename:")
```

```python
lower_diagonal = np.zeros(data.shape, dtype=np.uint8)
upper_diagonal = np.zeros(data.shape, dtype=np.uint8)

for r in range(y):
    for c in range(x):
        if r - m * c > 0:
            lower_diagonal[r, c] = data[r, c]
        else:
            upper_diagonal[r, c] = data[r, c]
```

```python
fig = plt.figure(figsize=(20,20))

rows = 1
cols = 2

i = 1

fig.add_subplot(rows, cols, 1)
plt.axis("off")
plt.title("Lower Diagonal")
plt.imshow(lower_diagonal)

fig.add_subplot(rows, cols, 2)
plt.axis("off")
plt.title("Upper Diagonal")
plt.imshow(upper_diagonal)

plt.savefig(file3)
plt.show()
```
