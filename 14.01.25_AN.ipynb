### Image border (outside)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file1 = input("Enter a rgb image filename:")
file2 = input("Enter output image filename:")
```

```python
data = plt.imread(file1)
```

```python
data.shape
```

```python
border_pixels = 10
```

```python
shape_x = data.shape[0] + 2 * border_pixels
shape_y = data.shape[1] + 2 * border_pixels

shape_x, shape_y
```

```python
img = np.zeros([shape_x, shape_y, data.shape[2]], dtype=np.uint8)
img.shape
```

```python
for r in range(data.shape[0]):
    for c in range(data.shape[1]):
        img[border_pixels+r, border_pixels+c] = data[r, c]
```

```python
rows = 1
cols = 2

fig = plt.figure(figsize=(20,20))

fig.add_subplot(rows, cols, 1)
plt.axis("off")
plt.title("No border")
plt.imshow(data)

fig.add_subplot(rows, cols, 2)
plt.axis("off")
plt.title("With border")
plt.imshow(img)

plt.savefig(file2)
plt.show()
```

### Image Border (inside)

```python
file3 = input("Enter output image filename:")
```

```python
img2 = np.zeros(data.shape, dtype=np.uint8)
```

```python
b = border_pixels
r, c = data.shape[:2]

img2[b: r - b, b: c - b] = data[b: r - b, b: c - b]
```

```python
rows = 1
cols = 2

fig = plt.figure(figsize=(20,20))

fig.add_subplot(rows, cols, 1)
plt.axis("off")
plt.title("No border")
plt.imshow(data)

fig.add_subplot(rows, cols, 2)
plt.axis("off")
plt.title("With border")
plt.imshow(img2)

plt.savefig(file3)
plt.show()
```

### Reflection

```python
file4 = input("Enter output image filename:")
```

```python
vflip = data[:,::-1]
hflip = data[::-1, :]
vhflip = data[::-1,::-1]
```

```python
rows = 2
cols = 2

fig = plt.figure(figsize=(20,20))

fig.add_subplot(rows, cols, 1)
plt.axis("off")
plt.title("Original")
plt.imshow(data)

fig.add_subplot(rows, cols, 2)
plt.axis("off")
plt.title("Vertical Flip")
plt.imshow(vflip)

fig.add_subplot(rows, cols, 3)
plt.axis("off")
plt.title("Horizontal Flip")
plt.imshow(hflip)

fig.add_subplot(rows, cols, 4)
plt.axis("off")
plt.title("Vertical Horizontal Flip")
plt.imshow(vhflip)

plt.savefig(file4)
plt.show()
```

### Split

```python
file5 = input("Enter output image filename:")
```

```python
lsplit = data[:,:c//2]
rsplit = data[:,c//2:]
usplit = data[:r//2,:]
dsplit = data[r//2:,:]
```
