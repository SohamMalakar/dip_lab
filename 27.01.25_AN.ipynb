### Scaling

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file1 = input("Enter a rgb image filename:")
file2 = input("Enter output image filename:")
```

```python
data = plt.imread(file1)
data.shape
```

```python
x = data.shape[0] * 2
y = data.shape[1] * 2
scaled = np.zeros((x, y) + data.shape[2:], dtype=np.uint8)
scaled.shape
```

```python
for i in range(x):
    for j in range(y):
        scaled[i, j] = data[i//2, j//2]
```

```python
plt.figure(figsize=(20, 20))

canvas = np.zeros((scaled.shape[0], 2 * scaled.shape[1], 3), dtype=np.uint8)

canvas[: data.shape[0], : data.shape[1]] = data
canvas[: scaled.shape[0], scaled.shape[1]: scaled.shape[1] + scaled.shape[1]] = scaled

plt.axis('off')
plt.imshow(canvas)
```
