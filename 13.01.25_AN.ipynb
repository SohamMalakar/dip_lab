```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file1 = input("Enter a rgb image filename:")
file2 = input("Enter output image filename:")
```

```python
img = plt.imread(file1)
img.shape
```

```python
r, g, b = img[:,:,0], img[:,:,1], img[:,:,2]

# convert rgb image to grayscale image
gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
```

### These next two works only with jpg images

```python
bw = gray.copy()

# convert grayscale image to black-and-white image
for r in range(bw.shape[0]):
    for c in range(bw.shape[1]):
        if bw[r, c] > 127:
            bw[r, c] = 255
        else:
            bw[r, c] = 0
```

```python
inversed = gray.copy()

# convert grayscale image to negative black and white image
for r in range(inversed.shape[0]):
    for c in range(inversed.shape[1]):
        if inversed[r, c] > 127:
            inversed[r, c] = 0
        else:
            inversed[r, c] = 255
```

```python
# create a collage of these 4 type of images
rows = 1
cols = 4

fig = plt.figure(figsize=(20,20))

fig.add_subplot(rows, cols, 1)
plt.axis("off")
plt.title("RGB image")
plt.imshow(img)

fig.add_subplot(rows, cols, 2)
plt.axis("off")
plt.title("Grayscale image")
plt.imshow(gray, cmap='gray')

fig.add_subplot(rows, cols, 3)
plt.axis("off")
plt.title("Black and white image")
plt.imshow(bw, cmap='gray')

fig.add_subplot(rows, cols, 4)
plt.axis("off")
plt.title("Negative black and white image")
plt.imshow(inversed, cmap='gray')

plt.savefig(file2)
plt.show()
```
